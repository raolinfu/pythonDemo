```python
%matplotlib notebook
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
csvFile1 = open("meet", "r")
reader1 = csv.reader(csvFile1)

result1 = []
for item in reader1:
    result1.append(int(item[0]))
    
csvFile2 = open("no_meet", "r")
reader2 = csv.reader(csvFile2)

result2 = []
for item in reader2:
    result2.append(int(item[0]))
    

csvFile3 = open("meet_count", "r")
reader3 = csv.reader(csvFile3)
result3 = []
for item in reader3:
    result3.append(int(item[0]))
    
csvFile4 = open("no_meet_count", "r")
reader4 = csv.reader(csvFile4)

result4 = []
for item in reader4:
    result4.append(int(item[0]))
```

```python
data1 = np.array(result1)
data2 = np.array(result2)
data3 = np.array(result3)
data4 = np.array(result4)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)
ax1.set_xlabel("query")
ax1.set_ylabel("ticks")
ax1.plot(data1, label = 'meet translated', color = 'blue')
ax1.plot(data2, label = 'need translation', color = 'red')
ax1.legend(loc='best')

ax2.set_xlabel("query")
ax2.set_ylabel("count")
ax2.plot(data3, label = 'meet translated', color = 'blue')
ax2.plot(data4, label = 'need translation', color = 'red')
ax2.legend(loc='best')
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)
plt.savefig('test.jpg')
```
